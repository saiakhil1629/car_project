```python
import pandas as pd
import numpy as np
# Load the dataset
dataset = pd.read_csv("/content/age_salary_dataset.csv")
#printing the dataset
dataset
```

```python
#Extracting independent varible(X) and dependent()
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
print('indepent varible (X):',X)
print('indepent varible (Y):',Y)
```

```python
# deling with missing data
from sklearn.impute import SimpleImputer
# Replacing the missing value using transform method
imp = SimpleImputer(missing_values=np.nan,strategy="mean")
X= imp.fit_transform(X)
Y= Y.reshape(-1,1)
Y=imp.fit_transform(Y)
Y=Y.reshape(-1)
print(X)
print(Y)
```

```python
#convert X and Y to DataFrames
X_df = pd.DataFrame(X, columns=dataset.columns[:-1])
Y_df = pd.DataFrame(Y, columns=[dataset.columns[-1]])
# Combine X and Y side by side using pandas.concat
combined_data = pd.concat([X_df,Y_df],axis=1)
# Display the combined data
print("combinet Data:")
print(combined_data)
```

```python
# convert the comined_data to a DataFrame
combined_df = pd.DataFrame(combined_data, columns=list(dataset.columns) )
# save the combined data to an Excel file
combined_df.to_excel("/content/combined_data.xlsx", index=False)
# Display the first few rows of the combined DataFrame
#print("combined DataFrame:")
#print(combined_df.head())
```
